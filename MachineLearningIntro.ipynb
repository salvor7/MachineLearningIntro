```python
from matplotlib import rc
%matplotlib inline
```

```python
rc("font", family="serif", size=36)
rc("text", usetex=False)
```

```python
import daft
```

```python
from collections import namedtuple
```

```python
Rectangle = namedtuple('Rectangle', 'x y w h')
```

```python

names = ['Statistics',
         'Data Mining',
         'Machine\nLearning',
        ]

offset = 0.4
background = 2*len(names) + offset, 2*len(names) + offset
max_w, max_h = [b - offset for b in background]

squares = [daft.PGM([*background], origin=[-offset/2, -offset/2]) for i in range(3)]

for idx, name in enumerate(names):
    rec = Rectangle(x=1.5*idx-idx*0.2, 
                    y=1.5*idx-idx*0.2, 
                    w=max_w-1.5*idx, 
                    h=max_h-1.5*idx
                   )
    
    for square in squares[:idx+1]:
        square.add_plate(daft.Plate(rec, label=name, label_offset=(4,10)))

for idx, square in enumerate(squares):
    square.render()
    square.figure.savefig("figures/subsetsss"+str(idx)+".png", dpi=100)
```

(This is true but dissatisfying.)


## Machine Learning Definition Take 2

A computer program is said to learn from experience $E$  
with respect to some class of tasks $T$  
and performance measure $P$  

if its performance at tasks in $T$,  
as measured by $P$,  
improves with experience $E$.

### Example - Patrick's Long Stay Predictor

$T$ - Classification of Long Stay (LOS > 30 days) versus not

$E$ - Inpatient data

$P$ - Mean Squared Error 

## Can be made precise by getting more precise about...

$T$ - Tasks

$E$ - Experience

$P$ - Performance
